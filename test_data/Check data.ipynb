```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.reshape(np.fromfile('filfile.fil', dtype=np.float32), (-1, 1024)).T
```

```python
for isamp in np.arange(128):
    print(data[1][isamp])
```

```python
fig = plt.figure(figsize=(15,10))

plt.imshow(data[1:, 1024:], aspect='auto')
```

```python
bandpass = np.sum(data[1:, :], axis=1)
```

```python
fig = plt.figure(figsize=(15,10))

plt.plot(bandpass)
```
